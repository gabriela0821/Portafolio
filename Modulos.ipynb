# **Modulo de entrada:**

```python
import wfdb
import numpy as np

pepito = wfdb.rdrecord("103")
print(pepito)
print(pepito.p_signal)

signals = pepito.p_signal
anotac = wfdb.rdann('103','atr').sample
wfdb.plot_wfdb(pepito)
```

```python
from pickleshare import print_function
import random

lista1=[] #lista de Vboser
signal1=signals[:,0]

for i in anotac:
  vb=0

  if i==649875:
    vb=signal1[i-32:]
  
  else:
    vb=signal1[i-16:i+16]

  lista1.append(vb)
print(lista1)
```

```python
lista2=[] #lista de no Vboser
signal1=signals[:,0]

for i in range(0, len(anotac)):
  vb=0

  if anotac[i]==649875:
    continue
  
  else:
    for j in range(0, 3):
      posA=random.randint(anotac[i]+30, anotac[i+1]-30)
      vb=signal1[posA-16:posA+16]
      lista2.append(vb)


print(lista2)
```

# **Modulo de salida:**

```python
import matplotlib.pyplot as plt
signals = pepito.p_signal
anotac = wfdb.rdann('103','atr').sample  #aqui se muestra donde esta el latido
signal1 = signals[:,0]
vb = signal1[anotac[1]-70: anotac[1]+70]#latidos que tienen un rango de hasta 140 muestras
print(vb)
```

```python
start=0
end=15
signal1=signals[:,0]
media=[]
desviacion=[]

while(end!=(len(signals)+7)):

  temp=signal1[start:end]
  if start==0:
      for i in range(0,15):
        media.append(np.mean(temp))
        desviacion.append(np.std(temp))

  else:
      for i in range(0,7):
        media.append(np.mean(temp))
        desviacion.append(np.std(temp))
  start+=7
  end+=7
print(len(media))
```

```python
plt.plot(media[:1000])
plt.plot(desviacion[:1000])
```
